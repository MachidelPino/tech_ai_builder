```python
from langchain_experimental.text_splitter import SemanticChunker
from langchain_huggingface import HuggingFaceEmbeddings

hf_emb_model = HuggingFaceEmbeddings(model_name ='intfloat/multilingual-e5-small')
semantic_hf_splitter = SemanticChunker(hf_emb_model)
semantic_hf_chunks = semantic_hf_splitter.split_documents(all_docs)
len(semantic_hf_chunks)
```

```python
import os
from dotenv import load_dotenv

from pinecone import Pinecone
from langchain_pinecone import PineconeVectorStore
from langchain_huggingface import HuggingFaceEmbeddings

# Fuerza la recarga del .env, aunque la variable ya existiera
load_dotenv(override=True)

INDEX_NAME = "langchain-rag"

pinecone_api_key = os.getenv("PINECONE_API_KEY")

if not pinecone_api_key:
    raise RuntimeError("No se encontró PINECONE_API_KEY en el archivo .env")

pinecone_api_key = pinecone_api_key.strip()

# 1. Cliente de Pinecone
pc = Pinecone(api_key=pinecone_api_key)

# 2. Obtener el host real del índice asociado a esta API key
description = pc.describe_index(name=INDEX_NAME)

host = (
    description.host
    if hasattr(description, "host")
    else description["host"]
)

print("Índice:", INDEX_NAME)
print("Host detectado:", host)

# 3. Crear conexión usando ese host, no uno copiado manualmente
index = pc.Index(host=host)

# 4. Probar autenticación directamente contra el Data Plane
stats = index.describe_index_stats()
print("Conexión correcta:", stats)

# 5. Embeddings
hf_emb_model = HuggingFaceEmbeddings(
    model_name="intfloat/multilingual-e5-small"
)

# 6. Vector store usando el objeto index ya autenticado
vector_store = PineconeVectorStore(
    index=index,
    embedding=hf_emb_model,
)

# async_req=False da un traceback más directo si algo falla
ids = vector_store.add_documents(
    documents=semantic_hf_chunks,
    async_req=False,
)

print(f"Se agregaron {len(ids)} documentos")
```

```python
from langchain_community.vectorstores import InMemoryVectorStore

vectorstore = InMemoryVectorStore.from_documents(
    documents=semantic_hf_chunks, embedding=hf_emb_model
)
```

```python
retriever = vectorstore.as_retriever(search_kwargs={'k':3})
retriever.invoke("Robo de la tarjeta")
```
