```python
import os
from dotenv import load_dotenv

load_dotenv()

GEMINI_API_KEY = os.getenv('GEMINI_API_KEY')
GROQ_API_KEY = os.getenv('GROQ_API_KEY')
PINECONE_API_KEY = os.getenv('PINECONE_API_KEY')
```

```python
from langchain_community.document_loaders import TextLoader

documentos = TextLoader(r'C:\Users\machi\Desktop\tech_ai_builder\temp\LANGCHAIN-RAG\documentos\mc_beneficios_global.txt',encoding='utf-8').load()
documentos
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(chunk_size= 1000, chunk_overlap=100)
fragmentos = splitter.split_documents(documentos)
```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings

modelo_embeddings = GoogleGenerativeAIEmbeddings(model = "gemini-embedding-001", google_api_key=GEMINI_API_KEY)
```

```python
from langchain_community.vectorstores import InMemoryVectorStore

vectorstore = InMemoryVectorStore.from_documents(documents=fragmentos, embedding = modelo_embeddings)
```

```python
retriever = vectorstore.as_retriever(search_kwargs={"k":2})
```

```python
retriever.invoke("robo de identidad")
```

```python
query = "Cómo debo proceder en el caso de que mi tarjeta haya sido robada?"

query_embed = modelo_embeddings.embed_query(query)
query_embed
```

```python
fragmentos_similares = retriever.invoke(query)
```

```python
fragmentos_similares
```

```python
fragmentos_similares = [fragmento.page_content for fragmento in fragmentos_similares]
fragmentos_similares
```

```python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate(
    [
        ("system","Responde utilizando exclusivamente el contenido que se anexa a continuación: \nContexto:\n{contexto}"),
        ("human","{query}")
    ]
)
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.output_parsers import StrOutputParser

modelo = ChatGoogleGenerativeAI(
    api_key = GEMINI_API_KEY,
    model = "gemini-3.1-flash-lite",
    temperature = 0.2
)
```

```python
modelo.invoke(query)
```

```python
cadena = prompt | modelo | StrOutputParser()

fragmentos = retriever.invoke(query)
contexto = "\n\n".join(fragmento.page_content for fragmento in fragmentos)

cadena.invoke({"query":query,"contexto":contexto})
```

```python
from langchain_core.globals import set_debug

set_debug(True)
cadena.invoke({"query":query,"contexto":contexto})
```
